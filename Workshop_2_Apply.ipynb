- Compute reactions at both supports

```python
import numpy as np
import matplotlib as plt
import matrixmethod as mm
%config InlineBackend.figure_formats = ['svg']
```

```python
EI = 1500
EA = 1000
q = 9
L = 5
phibar = 0.15

mm.Node.clear()
mm.Element.clear()

nodes = []

nodes.append(mm.Node(0,0))
nodes.append(mm.Node(L,-L))  # z is downwards - imp
nodes.append(mm.Node(2*L,0))

elems = []

elems.append(mm.Element(nodes[0], nodes[1]))
elems.append(mm.Element(nodes[1], nodes[2]))

section = {}
section['EI'] = EI
section['EA'] = EA

for elem in elems:
    elem.set_section(section)
    print(elem)  # section both EI and EA stiffness

con = mm.Constrainer()

con.fix_dof (nodes[0], 0) # x-axis displacement = 0
con.fix_dof (nodes[0], 1) # z-axis displacement = 0
con.fix_dof (nodes[2], 0) # x-axis displacement = 0
con.fix_dof (nodes[2], 1) # z-axis displacement = 0
con.fix_dof (nodes[2], 2, phibar) # # prescribed rotation / moment = phibar (+ve cause clockwise)

elems[0].add_distributed_load([0,q])  # local transverse = q
elems[1].add_distributed_load([0,2*q]) # local transverse = 2q 

print(con)

global_k = np.zeros ((3*len(nodes), 3*len(nodes))) # 3 dof x 3 nodes
global_f = np.zeros (3*len(nodes))

for e in elems:
    elmat = e.stiffness()
    idofs = e.global_dofs()
    
    global_k[np.ix_(idofs,idofs)] += elmat

for n in nodes:
    global_f[n.dofs] += n.p

Kc, Fc = con.constrain ( global_k, global_f )  # applying a Dirichlet BC 
u_free = np.matmul ( np.linalg.inv(Kc), Fc ) #u_free since not fixed
print(u_free)

print(con.support_reactions(global_k,u_free,global_f))
```

**Nodal Displacement & Support Reactions:**



```python
# Local Displacement: global_c = False
for elem in elems:
    u_elem = con.full_disp(u_free)[elem.global_dofs()]
    elem.plot_displaced(u_elem,num_points=51,global_c=False,scale=1)
```

```python
# Global Displacement, global_c = True
for elem in elems:
    u_elem = con.full_disp(u_free)[elem.global_dofs()]
    elem.plot_displaced(u_elem,num_points=51,global_c=True,scale=1)
```

```python
# Local Moment Lines: global_c = False
for elem in elems:
    u_elem = con.full_disp(u_free)[elem.global_dofs()]
    elem.plot_moment_diagram(u_elem,num_points=20,global_c=False)
```

```python
# Global Moment Lines: global_c = True
for elem in elems:
    u_elem = con.full_disp(u_free)[elem.global_dofs()]
    elem.plot_moment_diagram(u_elem,num_points=20,global_c=True,scale=0.05)
```
